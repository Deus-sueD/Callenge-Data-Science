#1. Análisis de facturación



```python
from matplotlib import pyplot as plt

def graficar(tienda):
  Tiendasportabla = tienda['Lugar de Compra']
  Tiendasunicasportabla = list(set(Tiendasportabla))

  ventastotales = []
  ventasportienda = []

  for x in Tiendasunicasportabla:
    #Agrega a una lista de tuplas los valores de la ciudad y ademas todos los valores de venta
    ventasportienda.append((x,tienda[tienda['Lugar de Compra'] == x]['Precio']))
    #Agrega a una lista de tuplas los valores de la ciudad y la suma de todas sus ventas
    ventastotales.append((x,float(tienda[tienda['Lugar de Compra'] == x]['Precio'].sum())))

  #Separamos la lista de tupplas con zip
  ciudad, venta = zip(*ventastotales)

  #definimos el tamaño de la grafica

  fig_width = max(10, len(ciudad))
  fig_height = 5
  plt.figure(figsize=(fig_width, fig_height))

  #Imprimimos grafica
  plt.bar(ciudad, venta)
  plt.title('Ventas por tienda')
  plt.xlabel('Tiendas')
  plt.ylabel('Ventas')

  #mostrar las etiquetas de x verticalmente
  plt.xticks(rotation=90)

  #ajusta el tamaño de las etiquetas

  #formatear el eje y para que muestre valores de miles(ambas formas son lo mismo)
  #plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
  plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"{int(x):,}"))
  plt.tight_layout()
  plt.show()

def Tiendaspordf(tienda):
  Tiendasunicasportabla = list(set(tienda['Lugar de Compra']))

  ventastotales = [(x,float(tienda[tienda['Lugar de Compra'] == x]['Precio'].sum())) for x in Tiendasunicasportabla]
  '''
  Si quereemos retornar de mayor a menor:
  return sorted(ventastotales, key=lambda x: x[1], reverse=True)
  '''
  #Retornamos de menor a mayor las ventas
  return sorted(ventastotales, key=lambda x: x[1])

def graficardemayoraminor(ventasxtienda,Titulodelgrafico,labelx,labely):
  ciudad, venta = zip(*ventasxtienda)
  fig_width = max(10, len(ciudad))
  fig_height = 5
  plt.figure(figsize=(fig_width, fig_height))
  plt.bar(ciudad, venta)
  plt.title(f'{Titulodelgrafico}')
  plt.xlabel(f'{labelx}')
  plt.ylabel(f'{labely}')
  plt.xticks(rotation=90)
  plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"{int(x):,}"))
  plt.tight_layout()
  plt.show()
  return

graficardemayoraminor(Tiendaspordf(tienda),'Ventas por zona de "Tienda 1"','Tiendas','Ventas')
graficardemayoraminor(Tiendaspordf(tienda2),'Ventas por zona de "Tienda 2"','Tiendas','Ventas')
graficardemayoraminor(Tiendaspordf(tienda3),'Ventas por zona de "Tienda 3"','Tiendas','Ventas')
graficardemayoraminor(Tiendaspordf(tienda4),'Ventas por zona de "Tienda 4"','Tiendas','Ventas')



```

# 2. Ventas por categoría

```python
def Ventasporcategoriaxtienda(tienda):
  Categoriaportabla = list(set(tienda['Categoría del Producto']))

  Categoriadelproductoportabla = list(tienda['Categoría del Producto'])
  contador = [Categoriadelproductoportabla.count(x) for x in Categoriaportabla]
  ventastotales = zip(Categoriaportabla, contador)

  #Retornamos de menor a mayor las ventas
  return sorted(ventastotales, key=lambda x: x[1])
```

```python
graficardemayoraminor(Ventasporcategoriaxtienda(tienda), 'Ventas por categoria de Tienda 1', 'Categoria','Ventas')
graficardemayoraminor(Ventasporcategoriaxtienda(tienda2), 'Ventas por categoria de Tienda 2', 'Categoria','Ventas')
graficardemayoraminor(Ventasporcategoriaxtienda(tienda3), 'Ventas por categoria de Tienda 3', 'Categoria','Ventas')
graficardemayoraminor(Ventasporcategoriaxtienda(tienda4), 'Ventas por categoria de Tienda 4', 'Categoria','Ventas')
```

# 3. Calificación promedio de la tienda


```python
def CalificacionPorTienda(tienda):
  Tiendasunicasportabla = list(set(tienda['Lugar de Compra']))

  Contador = [(float(tienda[tienda['Lugar de Compra'] == x]['Calificación'].count())) for x in Tiendasunicasportabla]
  Califtotales = {x : float(tienda[tienda['Lugar de Compra'] == x]['Calificación'].sum()) for x in Tiendasunicasportabla}

  tienda, sumadecalifiaciones = zip(*Califtotales.items())
  Califtotales = zip(tienda, [x/y for x,y in zip(sumadecalifiaciones,Contador)])
  #Retornamos de menor a mayor las ventas
  return sorted(Califtotales, key=lambda x: x[1])

```

```python
graficardemayoraminor(CalificacionPorTienda(tienda), 'Calificacion de ciudades para Tienda 1', 'Ciudad','Calificación')
graficardemayoraminor(CalificacionPorTienda(tienda2), 'Calificacion de ciudades para Tienda 2', 'Ciudad','Calificación')
graficardemayoraminor(CalificacionPorTienda(tienda3), 'Calificacion de ciudades para Tienda 3', 'Ciudad','Calificación')
graficardemayoraminor(CalificacionPorTienda(tienda4), 'Calificacion de ciudades para Tienda 4', 'Ciudad','Calificación')
```

# 4. Productos más y menos vendidos

```python
def Productosmastomenos(tienda):
  Productosunicosportabla = list(set(tienda['Producto']))
  Productosporseparado = list(tienda['Producto'])

  Contador = [Productosporseparado.count(x) for x in Productosunicosportabla]
  masvendidos = [(Productosunicosportabla[x], Contador[x]) for x in range(len(Productosunicosportabla))]

  '''
  Si quereemos retornar de mayor a menor:
  return sorted(ventastotales, key=lambda x: x[1], reverse=True)
  '''
  #Retornamos de menor a mayor las ventas
  return sorted(masvendidos, key=lambda x: x[1])
```

```python
graficardemayoraminor(Productosmastomenos(tienda), 'Productos mas vendidos de Tienda 1', 'Productos','Ventas')
graficardemayoraminor(Productosmastomenos(tienda2), 'Productos mas vendidos de Tienda 2', 'Productos','Ventas')
graficardemayoraminor(Productosmastomenos(tienda3), 'Productos mas vendidos de Tienda 3', 'Productos','Ventas')
graficardemayoraminor(Productosmastomenos(tienda4), 'Productos mas vendidos de Tienda 4', 'Productos','Ventas')
```

# 5. Envío promedio por tienda

```python
def EnvioPromedio(tienda):
  Tiendasunicasportabla = list(set(tienda['Lugar de Compra']))

  Contador = [(float(tienda[tienda['Lugar de Compra'] == x]['Costo de envío'].count())) for x in Tiendasunicasportabla]
  Enviototales = {x : float(tienda[tienda['Lugar de Compra'] == x]['Costo de envío'].sum()) for x in Tiendasunicasportabla}

  tienda, sumadeenvio = zip(*Enviototales.items())
  Enviototales = zip(tienda, [x/y for x,y in zip(sumadeenvio,Contador)])
  #Retornamos de menor a mayor las ventas
  return sorted(Enviototales, key=lambda x: x[1])
```

```python
EnvioPromedio(tienda)
EnvioPromedio(tienda2)
EnvioPromedio(tienda3)
EnvioPromedio(tienda4)
```

```python
graficardemayoraminor(EnvioPromedio(tienda), 'Envio promedio de Tienda 1', 'Ciudad','Costo promedio de Envio')
graficardemayoraminor(EnvioPromedio(tienda2), 'Envio promedio de Tienda 2', 'Ciudad','Costo promedio de Envio')
graficardemayoraminor(EnvioPromedio(tienda3), 'Envio promedio de Tienda 3', 'Ciudad','Costo promedio de Envio')
graficardemayoraminor(EnvioPromedio(tienda4), 'Envio promedio de Tienda 4', 'Ciudad','Costo promedio de Envio')
```

```python
def Promediodeenviototal(tienda, Nombre_de_la_tienda):
  TIENDA = EnvioPromedio(tienda)

  Promedioportienda = {Nombre_de_la_tienda : sum(x for _, x in TIENDA)}
  return Promedioportienda
```

```python
print(Promediodeenviototal(tienda,'TIENDA 1'))
print(Promediodeenviototal(tienda2,'TIENDA 2'))
print(Promediodeenviototal(tienda3,'TIENDA 3'))
print(Promediodeenviototal(tienda4,'TIENDA 4'))
```

```python
Costopromedioportiendas = []

TIENDA1 = Promediodeenviototal(tienda,'TIENDA 1')
TIENDA2 = Promediodeenviototal(tienda2,'TIENDA 2')
TIENDA3 = Promediodeenviototal(tienda3,'TIENDA 3')
TIENDA4 = Promediodeenviototal(tienda4,'TIENDA 4')

for s in [TIENDA1,TIENDA2,TIENDA3,TIENDA4]:
  for k,v in s.items():
    Costopromedioportiendas.append((k,v))

Costopromedioportiendas
```
