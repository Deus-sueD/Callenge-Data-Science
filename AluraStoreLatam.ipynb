#1. Análisis de facturación



```python
from matplotlib import pyplot as plt

def graficar(tienda):
  Tiendasportabla = tienda['Lugar de Compra']
  Tiendasunicasportabla = list(set(Tiendasportabla))

  ventastotales = []
  ventasportienda = []

  for x in Tiendasunicasportabla:
    #Agrega a una lista de tuplas los valores de la ciudad y ademas todos los valores de venta
    ventasportienda.append((x,tienda[tienda['Lugar de Compra'] == x]['Precio']))
    #Agrega a una lista de tuplas los valores de la ciudad y la suma de todas sus ventas
    ventastotales.append((x,float(tienda[tienda['Lugar de Compra'] == x]['Precio'].sum())))

  #Separamos la lista de tupplas con zip
  ciudad, venta = zip(*ventastotales)

  #definimos el tamaño de la grafica

  fig_width = max(10, len(ciudad))
  fig_height = 5
  plt.figure(figsize=(fig_width, fig_height))

  #Imprimimos grafica
  plt.bar(ciudad, venta)
  plt.title('Ventas por tienda')
  plt.xlabel('Tiendas')
  plt.ylabel('Ventas')

  #mostrar las etiquetas de x verticalmente
  plt.xticks(rotation=90)

  #ajusta el tamaño de las etiquetas

  #formatear el eje y para que muestre valores de miles(ambas formas son lo mismo)
  #plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
  plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"{int(x):,}"))
  plt.tight_layout()
  plt.show()

def Tiendaspordf(tienda):
  Tiendasunicasportabla = list(set(tienda['Lugar de Compra']))

  ventastotales = [(x,float(tienda[tienda['Lugar de Compra'] == x]['Precio'].sum())) for x in Tiendasunicasportabla]
  '''
  Si quereemos retornar de mayor a menor:
  return sorted(ventastotales, key=lambda x: x[1], reverse=True)
  '''
  #Retornamos de menor a mayor las ventas
  return sorted(ventastotales, key=lambda x: x[1])

def graficardemayoraminor(ventasxtienda,Titulodelgrafico,labelx,labely):
  ciudad, venta = zip(*ventasxtienda)
  fig_width = max(10, len(ciudad))
  fig_height = 5
  plt.figure(figsize=(fig_width, fig_height))
  plt.bar(ciudad, venta)
  plt.title(f'{Titulodelgrafico}')
  plt.xlabel(f'{labelx}')
  plt.ylabel(f'{labely}')
  plt.xticks(rotation=90)
  plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"{int(x):,}"))
  plt.tight_layout()
  plt.show()
  return

def Graficadepastel(tienda,Titulodelgrafico):
  import matplotlib.pyplot as plt

  Nombres = [nombre for nombre, _ in tienda]
  Ventas = [ventas for _, ventas in tienda]

  plt.pie(Ventas, labels=Nombres, autopct='%1.1f%%', startangle=90)
  plt.axis('equal')
  plt.title(Titulodelgrafico)
  plt.show()
```

```python
print(Tiendaspordf(tienda))
print(Tiendaspordf(tienda2))
print(Tiendaspordf(tienda3))
print(Tiendaspordf(tienda4))
```

# 2. Ventas por categoría

```python
def Ventasporcategoriaxtienda(tienda):
  Categoriaportabla = list(set(tienda['Categoría del Producto']))

  Categoriadelproductoportabla = list(tienda['Categoría del Producto'])
  contador = [Categoriadelproductoportabla.count(x) for x in Categoriaportabla]
  ventastotales = zip(Categoriaportabla, contador)

  #Retornamos de menor a mayor las ventas
  return sorted(ventastotales, key=lambda x: x[1])
```

```python
#graficardemayoraminor(Ventasporcategoriaxtienda(tiendasgral), 'Ventas por categoria', 'Categoria','Ventas')
Graficadepastel(Ventasporcategoriaxtienda(tiendasgral),'Ventas por categoria')
```

# 3. Calificación promedio de la tienda


```python
def CalificacionPorTienda(tienda):
  Tiendasunicasportabla = list(set(tienda['Lugar de Compra']))

  Contador = [(float(tienda[tienda['Lugar de Compra'] == x]['Calificación'].count())) for x in Tiendasunicasportabla]
  Califtotales = {x : float(tienda[tienda['Lugar de Compra'] == x]['Calificación'].sum()) for x in Tiendasunicasportabla}

  tienda, sumadecalifiaciones = zip(*Califtotales.items())
  Califtotales = zip(tienda, [x/y for x,y in zip(sumadecalifiaciones,Contador)])
  #Retornamos de menor a mayor las ventas
  return sorted(Califtotales, key=lambda x: x[1])

```

```python
graficardemayoraminor(CalificacionPorTienda(tiendasgral), 'Calificacion de ciudades', 'Ciudad','Calificación')
```

# 4. Productos más y menos vendidos

```python
def Productosmastomenos(tienda):
  Productosunicosportabla = list(set(tienda['Producto']))
  Productosporseparado = list(tienda['Producto'])

  Contador = [Productosporseparado.count(x) for x in Productosunicosportabla]
  masvendidos = [(Productosunicosportabla[x], Contador[x]) for x in range(len(Productosunicosportabla))]

  '''
  Si quereemos retornar de mayor a menor:
  return sorted(ventastotales, key=lambda x: x[1], reverse=True)
  '''
  #Retornamos de menor a mayor las ventas
  return sorted(masvendidos, key=lambda x: x[1])


```

# 5. Envío promedio por tienda

```python
def EnvioPromedio(tienda):
  Tiendasunicasportabla = list(set(tienda['Lugar de Compra']))

  Contador = [(float(tienda[tienda['Lugar de Compra'] == x]['Costo de envío'].count())) for x in Tiendasunicasportabla]
  Enviototales = {x : float(tienda[tienda['Lugar de Compra'] == x]['Costo de envío'].sum()) for x in Tiendasunicasportabla}

  tienda, sumadeenvio = zip(*Enviototales.items())
  Enviototales = zip(tienda, [x/y for x,y in zip(sumadeenvio,Contador)])
  #Retornamos de menor a mayor las ventas
  return sorted(Enviototales, key=lambda x: x[1])
```
